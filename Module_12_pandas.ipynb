### Transformation of df from inbuild data structure

```python
my_list = [['Alice',25],['Bob',30],['Charlie', 28]]
list_df = pd.DataFrame(my_list, columns=['Name', 'Age'], index=[1,2,3])
print(list_df)
print(type(list_df))
```

```python
my_tuple = (('Alice',25),('Bob',30),('Charlie',28))
tuple_df = pd.DataFrame(my_tuple, columns=['Name', 'Age'], index=[1,2,3])
print(tuple_df)
```

```python
my_dict = {
    'Name': ['Alice','Bob','Charlie'],
    'Age':[25,30,28]
} 
dict_df = pd.DataFrame(my_dict)
print(dict_df)
```

### Filtering Data Based on Condition

```python
# jara complete korche tader data niye ashbo
completed = df.loc[df['CompletionStatus'] == 'Completed']
completed
```

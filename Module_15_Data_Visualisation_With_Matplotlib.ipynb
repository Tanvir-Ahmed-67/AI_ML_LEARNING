```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 2D Line Plot

```python
hours = [2,3,4,1,5,7,3]
days = [1,2,3,4,5,6,7]
plt.plot(days,hours)
```

```python
# Title and Lebel adding
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Trend of a Student Study Time Over a Week')
plt.plot(days,hours)
```

```python
# Multiple data ploting
hours_1 = [2,3,4,1,5,7,3]
days = [1,2,3,4,5,6,7]
hours_2 = [3,5,6,2,7,9,5]
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Trend of a Student Study Time Over a Week')

plt.plot(days,hours_1)
plt.plot(days,hours_2)
```

### 2D Line from Pandas DF

```python
df = pd.read_csv("./Data_files/enrollment_data.csv")
df
```

```python
plt.xlabel('Year')
plt.ylabel('Count of Environment')
plt.title('Student Enrollment over the years')
plt.plot(df['Year'],df['Programming'])
```

```python
# adding labels using plt.legend()
plt.xlabel('Year')
plt.ylabel('Count of Environment')
plt.title('Student Enrollment over the years')

plt.plot(df['Year'],df['Programming'],label ='Programming')
plt.plot(df['Year'],df['Digital Marketing'], label = 'Digital Marketing')
plt.legend(loc='upper right')
```

```python
# adding some visualization
plt.xlabel('Year')
plt.ylabel('Count of Environment')
plt.title('Student Enrollment over the years')

plt.plot(df['Year'],df['Programming'],label ='Programming', color='red',linewidth=3, linestyle='dashed', marker='x', markersize = 7)
plt.plot(df['Year'],df['Digital Marketing'], label = 'Digital Marketing', color='black', linewidth=3, marker='o')
plt.legend(loc='upper left')
```

```python
# adding grid
plt.xlabel('Year')
plt.ylabel('Count of Environment')
plt.title('Student Enrollment over the years')

plt.plot(df['Year'],df['Programming'],label ='Programming', color='red',linewidth=3, linestyle='dashed', marker='x', markersize = 7)
plt.plot(df['Year'],df['Digital Marketing'], label = 'Digital Marketing', color='black', linewidth=3, marker='o')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

### Bar Chart

```python
df = pd.read_csv("./Data_files/student_data.csv")
df
```

```python
# Group By Completion Status
status = df.groupby('CompletionStatus').size()
#status
status.index
status.values
```

```python
# Bar chart Based on Completion Status
plt.xlabel('Completed Status')
plt.ylabel('Count')
plt.title('Bar Chart of Completed Status')
plt.bar(status.index, status.values, color=['red','green','yellow'], edgecolor='black')
plt.show()
```

```python
# Bar chart Based on Location Status
location = df.groupby('Location').size()

plt.xlabel('Location Of Students')
plt.ylabel('Count')
plt.title('Bar Chart of Completed Status')
plt.bar(location.index, location.values, color=['red','green','yellow','blue','grey'], edgecolor='black')
plt.show()
```

### Pie Chart

```python
status = df.groupby('CompletionStatus').size()
plt.pie(status,labels=status.index,autopct='%1.1f%%',explode = (0.1, 0, 0),shadow=True)
plt.show()
```

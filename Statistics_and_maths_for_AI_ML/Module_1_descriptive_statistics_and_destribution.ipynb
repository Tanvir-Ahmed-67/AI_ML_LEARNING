### Mean, Median, Mod

```python
import numpy as np
import pandas as pd
# simple Dataset (example: house price in thousands)
data = [120,125,128,122,126,124,600] # 600 is outlier
df = pd.DataFrame({'house_price':data})

# Mean (outlier sensetive)
mean = df['house_price'].mean()
print('Mean: ',mean)
```

```python
# Median (outlier resistance)
median = df['house_price'].median()
print('Median Value: ',median)
```

```python
# Mode
# adding some categorical values
df['city'] = ['dhaka', 'chittagont', 'rajshahi', 'dhaka', 'khulna', 'chittagong', 'dhaka']

mode = df['city'].mode()[0]
print('Mode Value: ',mode)
```

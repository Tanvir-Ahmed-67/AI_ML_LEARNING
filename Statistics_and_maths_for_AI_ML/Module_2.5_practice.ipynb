### Part A – Central Tendency and Spread

### Marks: 50, 60, 65, 70, 75, 80, 85, 90
### Find mean, median, mode.
### Replace 90 with 900 and recompute mean. Explain the effect of outliers.


```python
import numpy as np
import pandas as pd
marks = [50,60,65,70,75,80,85,90]
df = pd.DataFrame({'Marks':marks})
```

```python
mean = df['Marks'].mean()
print('Mean: ',mean)
```

```python
median = df['Marks'].median()
print('Median: ',median)
```

```python
mode = df['Marks'].mode()
print('Mode: ',mode)
```

```python
# Replacing 90 with 900 and calculating mean, median, mode
df['Marks'] = df['Marks'].replace(90,900)
df
```

```python
mean = df['Marks'].mean()
print('Mean: ',mean)
```

```python
median = df['Marks'].median()
print('Median: ',median)
```

```python
mode = df['Marks'].mode()
print('Mode: ',mode)
```

### Explanation of effects of outlier

when there was a value of 90, we got mean as 71.875, median as 72.5 and mode is calculated for categorical values. Since there is no categorical value so mode value is not important. Now when we changed 90 to 900 in dataframe,  we observed mean value is changed as 173.125 and median value changed as 72.5. In both case median value remains close or unchanged. So we can say, median value is outlier resustance but mean value is outlier sensitive.  

### Temperatures (°C): 29, 31, 33, 33, 32, 31, 30
### Calculate variance and standard deviation.
### Which measure is easier to interpret and why?


```python
temperature = [29,31,33,33,32,31,30]
df = pd.DataFrame({'Temperature':temperature})

mean = df['Temperature'].mean()
pop_variance = np.var(df['Temperature'], ddof=0) # population variance
sam_variance = np.var(df['Temperature'], ddof=1) # sample variance
pop_std = np.std(df['Temperature'], ddof= 0) # population SD
sam_std = np.std(df['Temperature'],ddof=1) # sample SD

print('Mean: ',mean)
print('Population Variance: ',pop_variance)
print('Sample Variance: ',sam_variance)
print('Population Standard Deviation: ',pop_std)
print('Sample Standard Deviation: ',sam_std)
```

standard deviation is easier to interept because it is single unit where variance is in square unit.

### Part B – Percentiles, IQR and Z-Score

### Cat weights (kg): 2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5
### Find P25, P50, P75 and IQR. Mark possible outliers.


```python
weight = [2.5,3.0,3.2,3.3,3.4,3.5,3.6,3.9,4.0,4.5]
#df = pd.DataFrame({'cat_weight': weight })
print('25 Percentile (Q1): ', np.percentile(weight, 25))
print('50 Percentile (Q2): ', np.percentile(weight, 50))
print('75 Percentile (Q3): ', np.percentile(weight, 75))
```

```python
Q3 = np.percentile(weight, 75)
Q1 = np.percentile(weight, 25)

IQR = Q3-Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print('IQR: ',IQR)
print('Outlier Range : ',lower_bound, "To", upper_bound)
```

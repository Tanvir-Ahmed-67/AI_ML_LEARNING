```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

### Scatter Plot 

```python
student = pd.read_csv('./Data_files/student_dataset_complete.csv')
student.head()
```

```python
student_marks = student[['Marks','study_hours','attendance_rate','gender']]
student_marks
```

```python
fig = px.scatter(student, x = 'time_study', y='Marks', color='gender', size='Tshirt_size', hover_data=['time_study','Marks','hostel'])
fig.show()
```

### Lineplot with plotly

```python
enrollment = pd.read_csv("./Data_files/enrollment_data.csv")
fig = px.line(enrollment, x = 'Year', y ='Programming')
fig.show()
```

```python
enrollment = pd.read_csv("./Data_files/enrollment_data.csv")
fig = px.line(enrollment, x = 'Year', y ='Digital Marketing', markers=True)
fig.show()
```

```python
fig.write_html('./Output_files/digital_marketing_data.html')
```

### Histogram using plotly

```python
fig = px.histogram(student_marks, x='attendance_rate',color='gender', nbins=5)
fig.show()
```

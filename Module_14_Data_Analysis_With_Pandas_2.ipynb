### Filtering Based On String

```python
# Finding in column - 'City' where 'new' words there. case sensetive
df.loc[df['City'].str.contains("New")]

# Ignoring the case
df.loc[df['City'].str.contains("new", case=False)]
```

```python
# Using regular expression. Starts with 'Los'
df.loc[df['City'].str.contains(r"^Los")]

# Using regular expression. Having 'Los'
df.loc[df['City'].str.contains(r"Los")]
```

```python
# Using regular expression. ending with 'rk'
df.loc[df['City'].str.contains(r"rk$")]
```

```python
# name starts with a vowel
df.loc[df['Name'].str.contains(r"^[AEIOU]")]
```

```python
# City name has New or Los
df.loc[df['City'].str.contains(r"New|Los")]
```

### Apply Function in DF

```python
df
```

```python
# min max scaling

mn = df['Total Marks'].min()
mx = df['Total Marks'].max()

# scaled marks
df['Scaled Marks'] = df['Total Marks'].apply(lambda x: (x-mn)/(mx-mn))
df
```

```python
# Custom build function

def grading_system(marks):
    if marks >= 260:
        return 'A+'
    elif marks >= 250:
        return 'A'
    else:
        return 'A-'

df['Grade'] = df['Total Marks'].apply(grading_system)
df
```

```python
# puro data frame er upor function apply kora
def marking_system(df):
    a = df['Data Structure Marks'] * 2
    b = df['Python Marks'] * 3
    c = df['Algorithm Marks'] * 4
    return a+b+c

df['Exceptional Marks'] = df.apply(marking_system, axis=1)
df
```

### DateTime and Timedelta

```python
import numpy as np
import pandas as pd
df = pd.read_csv("./Data_files/student_completed_data.csv")
df
```

```python
# getting days from datetime
df['Enrolment Day'] = df['Enrollment Date'].dt.day
df
```

```python
# getting month from datetime
df['Enrolment Month'] = df['Enrollment Date'].dt.month
df
```

```python
df['Finish Date'] = pd.to_datetime(df['FinishedDate'])
```

```python
df['Total time taken to finish'] = df['Finish Date'] - df['Enrollment Date']
```

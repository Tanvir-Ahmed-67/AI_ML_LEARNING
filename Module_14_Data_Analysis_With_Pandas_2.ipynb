```python
import pandas as pd 
data = {
    "Name":["Alice","Bob","Charli","David","Eve","Frank","Grace","Hanna","Sakib"],
    "City":["New York","Los Angeles","New York","Boston","New Delhi","Chicago","Michigan","Florida","H Los"],
    "Departmment":["HR","IT","FINANCE","IT","FINANCE","MARKETING","FINANCE","HR","IT"],
    "Salary":[50000,60000,550000,70000,52000,58000,62000,51000,53000]
}
df = pd.DataFrame(data)
df
```

### Filtering Based On String

```python
# Finding in column - 'City' where 'new' words there. case sensetive
df.loc[df['City'].str.contains("New")]

# Ignoring the case
df.loc[df['City'].str.contains("new", case=False)]
```

```python
# Using regular expression. Starts with 'Los'
df.loc[df['City'].str.contains(r"^Los")]

# Using regular expression. Having 'Los'
df.loc[df['City'].str.contains(r"Los")]
```

```python
# Using regular expression. ending with 'rk'
df.loc[df['City'].str.contains(r"rk$")]
```

```python
# name starts with a vowel
df.loc[df['Name'].str.contains(r"^[AEIOU]")]
```

```python
# City name has New or Los
df.loc[df['City'].str.contains(r"New|Los")]
```

### Imputing missing values for AGE and Embarked column
we will use median as age features were right skewed

```python
# Hnadeling neumeric missing values: Age 

age_median = df_titanic["Age"].median()
df_titanic["Age"] = df_titanic["Age"].fillna(age_median)

# Hnadeling neumeric missing values: Embarked
embarked_mode = df_titanic["Embarked"].mode()[0]
df_titanic["Embarked"] = df_titanic["Embarked"].fillna(embarked_mode)

# Drop Cabin (too many missing values)
df_titanic = df_titanic.drop(columns=["Cabin"])
```

```python
print("Missing values per column after handling: ")
df_titanic.isnull().sum()
```

```python
# Encoding Categorical Variables (Heart Dataset)
from sklearn.preprocessing import LabelEncoder
df_heart = pd.read_csv("../Data_files/heart.csv")
print("First Five rows of heart data set")
display(df_heart.head())

print("\n Columns data types: ")
display(df_heart.dtypes)
```

```python
# Categorical Features Exploration 
categorical_cols = ["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]
for c in categorical_cols:  
    plt.figure(figsize=(5,4))
    df_heart[c].value_counts().plot(kind="bar")
    plt.title(f"Value Counts for {c}")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```

### where there are binary categorical values we will do lebel encodin
### where there are more than two categorical values, we will do one hop encoding

```python
# Label Encoding for binary categorical column -> Sex and ExerciseAngina

le = LabelEncoder()
df_heart["Sex"] = le.fit_transform(df_heart["Sex"])
df_heart["ExerciseAngina"] = le.fit_transform(df_heart["ExerciseAngina"])
```

```python
df_heart.head(10)
```

```python
# Label Encoding for nominal categorical column
cat_cols = ["ChestPainType", "RestingECG", "ST_Slope"]
df_heart_encoded = pd.get_dummies(
    df_heart,
    columns= cat_cols,
    dtype=int
)
```

```python
df_heart_encoded.head(10)
```

### Normalization and Scaling 
scaling should only done on training dataset, not test dataset

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler 

# Assume df_heart_encoded is our working dataframe with target 'HeartDisease'

target_col = "HeartDisease"
X = df_heart_encoded.drop(columns=[target_col])
Y = df_heart_encoded[target_col]

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.25, random_state=42)

# Standard Scaling
scaler_sd = StandardScaler()
X_train_std = scaler_sd.fit_transform(X_train)  # Mean and SD will be calculated from X_train
X_test_std = scaler_sd.transform(X_test)


# Converts scaled arrays back to DataFrame for better visualization with column names
X_train_std_df = pd.DataFrame(X_train_std, columns = X_train.columns, index = X_train.index)

print("First five rows of X_train(Standard Scaling)")
X_train_std_df.head(5)
```

```python
# Scaling of test data
X_test_std_df = pd.DataFrame(X_test_std, columns = X_test.columns, index = X_test.index)
X_test_std_df.head(5)
```

```python
# MinMax Scaling
scaler_mm = MinMaxScaler()
X_train_mm = scaler_mm.fit_transform(X_train)
X_test_mm = scaler_mm.transform(X_test)

print("\n Displaying Standard Scaled Data")
X_train_mm_df = pd.DataFrame(X_train_mm, columns = X_train.columns, index = X_train.index)
X_test_mm_df = pd.DataFrame(X_test_mm, columns = X_test.columns, index = X_test.index)

print("Display first 5 rows of X_train (Minmax Scaled):")
X_train_mm_df.head(5)
```
